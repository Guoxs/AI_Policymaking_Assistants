```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
state_i = 'mississippi'
test_state = 'texas'

def get_inflow_df(path, state_i):
    with open(path, 'rb') as f:
        policy  = pickle.load(f)
    state_list = ['arizona','mississippi', 'new mexico', 'texas', 'virginia']
    state_list.remove(state_i)
    inflow_dict = {state: [] for state in state_list}
    for i in range(len(policy)):
        for state in state_list:
            if state == state_i:
                continue
            inflow = policy[i].outflow[state]
            inflow_dict[state].append(inflow)
    inflow_df = pd.DataFrame(inflow_dict)
    return inflow_df


#####time period
plt.figure(figsize=(12,6))
base_path = f'outputs\\5 states\\6 weeks\\no_action_2025-12-17-20-07-44\mobility\\{state_i}_mobility.pkl'
base_inflow_df = get_inflow_df(base_path, state_i)
base_inflow_df[test_state].plot(color='orange', label='Ground Truth')

# 42mark
interval = 42
x_vals = base_inflow_df.index[::interval]
y_vals = base_inflow_df[test_state].iloc[::interval]
plt.scatter(x_vals, y_vals, color='orange', marker='o', s=40)

# gpt5_path = f'outputs\\5 states\\6 weeks\gpt-5-mini_2025-12-17-20-17-40\mobility\\{state_i}_mobility.pkl'
# gpt5_inflow_df = get_inflow_df(gpt5_path, state_i)
# gpt5_inflow_df[test_state].plot(color='green', label='gpt-5')

# gpt3_5_path = f"outputs\\5 states\\6 weeks\gpt-3.5-turbo-0125_2025-12-16-13-39-47-repo\\mobility\{state_i}_mobility.pkl"
# gpt3_5_inflow_df = get_inflow_df(gpt3_5_path, state_i)
# gpt3_5_inflow_df[test_state].plot(color='purple', label='LLM Agent')
# interval = 42
# x_vals = gpt3_5_inflow_df.index[::interval]
# y_vals = gpt3_5_inflow_df[test_state].iloc[::interval]
# plt.scatter(x_vals, y_vals, color='purple', marker='o', s=40)

# gpt3_5_v2_path = f"outputs\\5 states\\6 weeks\gpt-3.5-turbo-0125_2025-12-18-08-36-51\\mobility\{state_i}_mobility.pkl"
# gpt3_5_v2_inflow_df = get_inflow_df(gpt3_5_v2_path, state_i)
# gpt3_5_v2_inflow_df[test_state].plot(color='yellow', label='gpt-3.5-v2')

# rule_path = f'outputs\\5 states\\6 weeks\\rule_based_2025-12-17-20-46-30\mobility\\{state_i}_mobility.pkl'
# rule_inflow_df = get_inflow_df(rule_path, state_i)
# rule_inflow_df[test_state].plot(color='blue', label='rule')

# gpt4o_mini_path = f"outputs\\5 states\\6 weeks\\gpt-4o-mini_2025-12-18-00-10-38\\mobility\{state_i}_mobility.pkl"
# gpt4o_mini_inflow_df = get_inflow_df(gpt4o_mini_path, state_i)
# gpt4o_mini_inflow_df[test_state].plot(color='red', label='gpt-4o-mini')


gpt41_mini_path = f"outputs\\5 states\\6 weeks\\gpt-4.1-mini\gpt-4.1-mini_2025-12-19-22-04-40\\mobility\{state_i}_mobility.pkl"
gpt41_mini_inflow_df = get_inflow_df(gpt41_mini_path, state_i)
gpt41_mini_inflow_df[test_state].plot(color='red', label='LLM Agent')




#     base_policy  = pickle.load(f)
# state_list = ['arizona','mississippi', 'new mexico', 'texas', 'virginia']
# state_list.remove(state_i)
# base_inflow_dict = {state: [] for state in state_list}
# for i in range(len(base_policy)):
#     for state in state_list:
#         if state == state_i:
#             continue
#         inflow = base_policy[i].outflow[state]
#         base_inflow_dict[state].append(inflow)
# base_inflow_df = pd.DataFrame(base_inflow_dict)
# base_inflow_df[test_state].plot(color='orange', label='base')


# with open(f'outputs\\5 states\\6 weeks\load_policy_2025-12-17-20-17-40\mobility\\{state_i}_mobility.pkl', 'rb') as f:
#     result_policy  = pickle.load(f)
# inflow_dict = {state: [] for state in state_list}
# for i in range(len(result_policy)):
#     for state in state_list:
#         if state == state_i:
#             continue
#         inflow = result_policy[i].outflow[state]
#         inflow_dict[state].append(inflow)
# inflow_df = pd.DataFrame(inflow_dict)
# inflow_df[test_state].plot(color='green', label='load_policy')


# with open(f'outputs\\5 states\\6 weeks\\rule_based_2025-12-17-20-46-30\mobility\\{state_i}_mobility.pkl', 'rb') as f:
#     result_policy  = pickle.load(f)
# rule_dict = {state: [] for state in state_list}
# for i in range(len(result_policy)):
#     for state in state_list:
#         if state == state_i:
#             continue
#         inflow = result_policy[i].outflow[state]
#         rule_dict[state].append(inflow)
# rule_df = pd.DataFrame(rule_dict)
# rule_df[test_state].plot(color='blue', label='rule')


# with open(f"outputs\\5 states\\6 weeks\gpt-3.5-2025-12-17-20-52-43\\mobility\{state_i}_mobility.pkl", 'rb') as f:
#     result_policy  = pickle.load(f)
# gpt_dict = {state: [] for state in state_list}
# for i in range(len(result_policy)):
#     for state in state_list:
#         if state == state_i:
#             continue
#         inflow = result_policy[i].outflow[state]
#         gpt_dict[state].append(inflow)
# gpt_dict_df = pd.DataFrame(gpt_dict)
# gpt_dict_df[test_state].plot(color='purple', label='gpt-3.5')


# with open(f"outputs\\5 states\\6 weeks\gpt-3.5-2025-12-17-23-54-42\\mobility\{state_i}_mobility.pkl", 'rb') as f:
#     result_policy  = pickle.load(f)
# gpt_dict_2 = {state: [] for state in state_list}
# for i in range(len(result_policy)):
#     for state in state_list:
#         if state == state_i:
#             continue
#         inflow = result_policy[i].outflow[state]
#         gpt_dict_2[state].append(inflow)
# gpt_dict_df_2 = pd.DataFrame(gpt_dict)
# gpt_dict_df_2[test_state].plot(color='yellow', label='gpt-3.5-2')

# x_vals
for i in range(len(x_vals)):
    plt.axvline(x=x_vals[i], color='gray', linestyle='--', alpha=0.5)
    if i < len(x_vals) - 1:
        plt.axvspan(x_vals[i], x_vals[i+1], color=('lightblue' if i % 2 == 0 else 'lightyellow'), alpha=0.2)
plt.legend()
```
